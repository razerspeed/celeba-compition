```python
#We want to detect gender and youngness within an image 
#Hence we extracted the feature columns

important_feature=df[['image_id','Male','Young']].copy()
important_feature.replace(-1,0,inplace=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
   
fig = plt.figure(figsize = (10, 5))
values=important_feature['Male'].value_counts().values
# creating the bar plot
plt.bar(['Female','Male'], values,
        width = 0.4)
 
plt.xlabel("Gender Types")
plt.ylabel("No. of persons")
plt.title("Count of Genders")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
   
fig = plt.figure(figsize = (10, 5))
values=important_feature['Young'].value_counts().values
# creating the bar plot
plt.bar(['Young','Old'], values,
        width = 0.4)
 
plt.xlabel("Types of people")
plt.ylabel("No. of persons")

plt.show()

```
